```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Read movie box office data table file
file_path = "blockbusters.csv"  
data = pd.read_csv(file_path)
```

```python
# Select movie duration column
movie_duration = data['length']
# Perform standardization
scaler = StandardScaler()
# Perform standardization
scaled_duration = scaler.fit_transform(movie_duration.values.reshape(-1, 1))
```

```python
# Draw a histogram
plt.hist(scaled_duration, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Normalized movie duration distribution')
plt.xlabel('Standardized movie duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
